## SuperDataScience Podcast NLP Analysis

### Notebook# 1: Websocket_URL_Extraction

This notebook contains code for extracting the episode names for SuperDataScience podcast from September, 2016 to May, 2023. Create list of episodes URL for web scraping. On inspecting website https://www.superdatascience.com/podcast noticed that request is sent through websocket. Exported websocket as .har file and converted it into .txt file.

```python
import pandas as pd
import re
```

```python
with open ('../data/www.superdatascience.com.txt') as file: 
    text = file.read()
```

```python
re.findall(r'\"/podcast/([0-9a-zA-Z\-]+)\\\"', text)
```

```python
len(re.findall(r'\"/podcast/([0-9a-zA-Z\-]+)\\\"', text))
```

```python
SDS = (re.findall(r'\"/podcast/([0-9a-zA-Z\-]+)\\\"', text))
```

```python
SDS = list(set(SDS))
```

```python
len(SDS)
```

```python
# Function to URL links
def SDSLinks(url_list):
    sdslinks = []
    baseURL = 'https://www.superdatascience.com/podcast/'
    
    for url in url_list:
        sdslinks.append(baseURL + url)
    return sdslinks   
        
```

```python
links = SDSLinks(SDS)
```

```python
SDS_links = pd.DataFrame(links, columns=['url_link'])
```

```python
SDS_links['url_link'].loc[0]
```

```python
SDS_links.to_csv('../data/sds_url_links.csv', index = False)
```
