```python
def preprocess_text(text):
    doc = nlp(text)
    tokens = [token.lemma_.lower().strip() for token in doc if not token.is_stop
             and not token.is_digit
             and not token.is_punct
             and not token.is_space]
    return tokens
```

## Entity Detection

```python
sds = pd.read_csv('../data/sds_cleaned.csv')
```

```python
sds['guest_info'] = sds['guest_info'].str.replace('Subscribe on Website, Apple Podcasts, Spotify, Stitcher Radio or Tune', '')
```

```python
def process_text(row):
    guest_text = []
    text = row['guest_info']
    episode_no = row['episode_number']
    
    start_index = 1
    end_index = 100
    words = text.split()

    selected_text = words[start_index:end_index]

    # Join the selected_text as string
    guest_text = ' '.join(selected_text)
    
    # Create a DataFrame from the split text and speaker
    text_df = pd.DataFrame({'episode_number': [episode_no] , 'guest': [guest_text]})
    
    return text_df
```

```python
# Loop to process each row of sds_datascience dataframe
clean_guest = pd.DataFrame(columns=['episode_number', 'guest'])
for i, row in sds.iterrows():
    cleaned_text = process_text(row)
    clean_guest = clean_guest.append(cleaned_text, ignore_index = False)
```

```python
sds = pd.merge(sds, clean_guest, on = 'episode_number')
```

```python
def entity_detection_text(text):
    doc = nlp(text)
    entity = displacy.render(doc, style = 'ent')
    return entity
```

```python
sds['guest'].apply(entity_detection_text)
```

```python
def entity_detection(text):
    doc = nlp(text)
    entities = [(entity.text, entity.label_) for entity in doc.ents if entity.label_ in ['ORG', 'PERSON', 'GPE']]
    return entities
```

```python
sds['guest_entity'] = sds['guest'].apply(entity_detection)
```
