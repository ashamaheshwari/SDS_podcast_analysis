```python
# Function to URL links
def SDSLinks(url_list):
    sdslinks = []
    baseURL = 'https://www.superdatascience.com/podcast/'
    
    for url in url_list:
        sdslinks.append(baseURL + url)
    return sdslinks   
        
```
